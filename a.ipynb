```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from collections import Counter
from konlpy.tag import Okt
from PIL import Image
import numpy as np

with open('test.txt', 'r', encoding='utf-8') as f:
    text = f.read()
    
im = Image.open('image/다운로드.png')
mask_arr = np.array(im)

okt = Okt()
nouns = okt.nouns(text) # 명사만 추출

words = [n for n in nouns if len(n) > 1] # 단어의 길이가 1개인 것은 제외

c = Counter(words) # 위에서 얻은 words를 처리하여 단어별 빈도수 형태의 딕셔너리 데이터를 구함

wc = WordCloud(font_path='image/Font/malgunbd.ttf', width=400,mask = mask_arr, height=400, scale=2.0, max_font_size=250,
               background_color ='white')
gen = wc.generate_from_frequencies(c)
plt.figure()
plt.axis('off')
plt.imshow(gen)
```

```python
def generate_wordcloud(file_path, image_path):
    with open(file_path, 'r', encoding='utf-8') as f:
        text = f.read()

    im = Image.open(image_path)
    mask_arr = np.array(im)

    okt = Okt()
    nouns = okt.nouns(text)  # Extract only nouns

    words = [n for n in nouns if len(n) > 1]  # Exclude words with a length of 1

    c = Counter(words)  # Count the frequency of each word

    wc = WordCloud(font_path='/home/oh/workspace/Font/malgunbd.ttf', width=400, mask=mask_arr, height=400, scale=2.0,
                   max_font_size=250, background_color='white')
    gen = wc.generate_from_frequencies(c)

    plt.figure()
    plt.axis('off')
    plt.imshow(gen)
    plt.show()
```

```python
generate_wordcloud('test.txt', '/home/oh/workspace/다운로드.png')
```
